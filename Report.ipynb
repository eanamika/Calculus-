Gradient Descent for X**2

```python
from sympy import * 

# Inputs are: f(x), Xo, Tolerence, step_size
x = Symbol("x")
f = x**2     
Xo = 0.2
tolerence = 0.01
step_size = 0.01

# Initialization with: f'(x), tol>tolerence, Xn_1 = Xo
# where tol= |Xn - Xn_1|
dif = diff(f, x)
tol = 1
Xn_1 = Xo
iter= 0
while abs(tol)>tolerence:
    Xn = Xn_1 - dif.subs(x, Xn_1)*(step_size)
    tol =  Xn - Xn_1
    Xn_1 = Xn
    iter= iter+1
print("minima at x is:", float(Xn_1))
value = f.subs(x, Xn_1)
print("number of iteration is:", iter)
```

Newton-Raphson Method for X**2

```python
# Inputs are: f(x), Xo, Tolerence
x = symbols("x")
f = x**2       
Xo = 0.2
tolerence = 0.01
# initialization  with:
g = diff(f, x)
diffg = diff(g, x)
Xn_1 = Xo
tol = 1
iter = 0
while abs(tol)>tolerence:
    Xn = Xn_1 - g.subs(x, Xn_1)/diffg.subs(x, Xn_1)
    tol = Xn - Xn_1
    Xn_1 = Xn
    iter = iter+1
print("minima at x=", Xn_1)
value = f.subs(x, Xn_1)
print("number of iteration:", iter)
```

Gradient Descent for exp(x)-log(x)

```python
from sympy import *

# Inputs are: f(x), Xo, step_size, tolerance
x = Symbol("x")
f = exp(x)- log(x)     
Xo = 0.2
step_size = 0.01
tolerence = 0.01

# Initialization with: f'(x), tol>tolerence, Xn_1 = Xo
# where tol= |Xn - Xn_1|
dif = diff(f, x)
Xn_1 = Xo
iter= 0
tol = 1

while abs(tol)>tolerence:
    Xn = Xn_1 - dif.subs(x, Xn_1)*(step_size)
    tol =  Xn - Xn_1
    Xn_1 = Xn
    iter= iter+1
print("Minima at x is:", Xn_1)
value = f.subs(x, Xn_1)
print("Number of iterations are:", iter)
```
